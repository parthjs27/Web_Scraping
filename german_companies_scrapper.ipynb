```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Germany'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
company_titles = table.find_all('th')
print(company_titles)
```

```python
notable_titles = [company.text.strip() for company in company_titles[:-2]]
print(notable_titles)
```

```python
df = pd.DataFrame(columns=notable_titles)
df
```

```python
df = df.rename(columns={'Status': 'Status (P=Private, S=State; A=Active, D=Defunct)'})
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    individual_row_data[-2:] = [' | '.join(individual_row_data[-2:])]
    length = len(df)
    df.loc[length] = individual_row_data
    # print(individual_row_data)
```

```python
df
```

```python
df.to_csv(r'Notable_companies_in_Germany.csv', index=False)
```
